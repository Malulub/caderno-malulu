```python
import folium

#definir as coordenadas de porto alegre
latitude1 = -30.0346
longitude1 = -51.2177

#definir as coordenadas de SP
latitude2 = -23.5890
longitude2 = -46.6592

#definir as coordenadas de RJ
latitude3 = -22.9346
longitude3 = -43.4801

#definir as coordenadas de CR
latitude4 = -25.4191
longitude4 = -49.3071

#definir as coordenadas de BH
latitude5 = -19.8528
longitude5 = -43.9954


#Criando um mapa centrado em porto alegre
mapa = folium.Map(location=[latitude1, longitude1], zoom_start=12)


#adicionando um marcador para geral, vulgo pontinhos
folium.Marker([latitude1,longitude1], popup='Mercado Público, Porto Alegre, RS, Brasil').add_to(mapa)
folium.Marker([latitude2,longitude2], popup='Parque Ibirapuera, São Paulo, SP, Brasil').add_to(mapa)
folium.Marker([latitude3,longitude3], popup='Parque Estadual Pedra Branca, Rio de Janeiro, RJ, Brasil').add_to(mapa)
folium.Marker([latitude4,longitude4], popup='Parque Barigui, Curitiba, CR, Brasil').add_to(mapa)
folium.Marker([latitude5,longitude5], popup='Parque Francisco Lins, Belo Horizonte, BH, Brasil').add_to(mapa)

#exibindo mapa
mapa
```
